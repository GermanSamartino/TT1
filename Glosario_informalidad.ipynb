**NOTAS:**

- Codificación Oculta del INDEC (-9): Cuando la gente decide no declarar sus ingresos, el INDEC no deja la celda vacía (NaN), sino que le asigna el valor -9. Si no se limpia esto (como lo agregué en la función procesar_informalidad), cuando apliques el logaritmo de ingresos para tu modelo, te va a arrojar error matemático o te va a sesgar violentamente las medias hacia abajo.

- Armonización de Trimestres: Elegí el cuarto trimestre (4) para cada hito. Esto es una buena práctica estadística para eliminar el sesgo de la "estacionalidad laboral" (por ejemplo, el empleo temporal de verano o el pago de aguinaldos que impacta fuertemente en el primer o segundo trimestre).

# Glosario de variables

### 1. Claves de Identificación (IDs)
Estas variables son exclusivas para la identificación y el cruce de bases (como Personas con Hogares), no se incluyen como nodos en el algoritmo.

*   **`CODUSU`**: Código de identificación de la vivienda. Es un string alfanumérico único para cada vivienda encuestada.
*   **`NRO_HOGAR`**: Número de hogar dentro de la vivienda. Suele ser 1, salvo en casos donde conviven distintos núcleos familiares con presupuestos separados bajo el mismo techo.
*   **`COMPONENTE`**: Número de orden de la persona dentro del hogar (1 suele ser el jefe/a de hogar). La llave primaria de cada individuo se arma concatenando `CODUSU` + `NRO_HOGAR` + `COMPONENTE`.

### 2. Variables Temporales
*   **`ANO4`**: Año de relevamiento (ej. 2019, 2021, 2025).
*   **`TRIMESTRE`**: Trimestre del año en el que se realizó la encuesta (1, 2, 3 o 4).

### 3. Filtros y Target (Condición Laboral)
*   **`ESTADO`**: Condición de actividad de la persona.
    *   0: Entrevista individual no realizada
    *   1: Ocupado (Filtro para tu universo de análisis)
    *   2: Desocupado
    *   3: Inactivo (Jubilados, estudiantes que no buscan trabajo, etc.)
    *   4: Menor de 10 años
*   **`CAT_OCUP`**: Categoría ocupacional.
    *   1: Patrón
    *   2: Cuenta propia
    *   3: Obrero o empleado (Asalariado - Segundo filtro para tu universo)
    *   4: Trabajador familiar sin remuneración
    *   9: Ns/Nr
*   **`PP07H`**: Target de Informalidad. Responde a la pregunta: *¿Por ese trabajo le descuentan para la jubilación?*
    *   1: Sí (Trabajo Formal)
    *   2: No (Trabajo Informal)

### 4. Nodos Demográficos y Educativos
*   **`CH04`**: Sexo.
    *   1: Varón
    *   2: Mujer
*   **`CH06`**: Edad en años cumplidos. (Numérica continua).
*   **`CH07`**: Estado civil actual.
    *   1: Unido/a
    *   2: Casado/a
    *   3: Separado/a o divorciado/a
    *   4: Viudo/a
    *   5: Soltero/a
*   **`NIVEL_ED`**: Nivel educativo alcanzado. (Ordinal).
    *   1: Primaria Incompleta (incluye educación especial)
    *   2: Primaria Completa
    *   3: Secundaria Incompleta
    *   4: Secundaria Completa
    *   5: Superior Universitaria Incompleta
    *   6: Superior Universitaria Completa
    *   7: Sin instrucción
    *   9: Ns/Nr

### 5. Nodos Estructurales y Económicos
*   **`REGION`**: Código de la región estadística.
    *   1: Gran Buenos Aires (CABA y GBA)
    *   40: NOA (Noroeste)
    *   41: NEA (Noreste)
    *   42: Cuyo
    *   43: Pampeana
    *   44: Patagonia
*   **`PP04A`**: Tamaño del establecimiento. Responde a: *El negocio, empresa o institución, ¿cuántas personas ocupadas tiene en total?*
    *   1: 1 persona
    *   2: 2 a 5 personas
    *   3: 6 a 40 personas
    *   4: 41 a 100 personas
    *   5: 101 a 500 personas
    *   6: Más de 500 personas
    *   9: Ns/Nr
*   **`PP03D`**: Cantidad de ocupaciones. Mide el nivel de pluriempleo de la persona.
    *   1: Una sola ocupación
    *   2: Más de una ocupación
*   **`PP04G`**: Lugar de Trabajo.
    *   11 = En un local/oficina/establecimiento/negocio/taller
    *   12 = En esta vivienda (con lugar exclusivo)
    *   13 = En una chacra/finca
    *   2 = En puesto o kiosco fijo callejero
    *   3 = En vehículos: bicicleta, moto, auto, barco, bote (no incluye servicio de transporte)
    *   4 = En vehículo para transporte de personas y mercaderías –aéreo,
marítimo, terrestre (incluye taxis, colectivos, camiones, furgones,
transporte de combustible, mudanzas, etc.)–
    *   5 = En obras en construcción, de infraestructura, minería o similares
    *   6 = En esta vivienda (sin lugar exclusivo)
    *   7 = En la vivienda del socio o del patrón
    *   8 = En el domicilio/local de los clientes
    *   9 = En la calle, espacios públicos, ambulante, de casa en casa, puesto
móvil callejero
    *   10 = En otros lugares (especificar)
*   **`PP04B_COD`**: Rama de Actividad. Código de la ocupación principal del establecimiento según el CAES (Clasificador de Actividades Económicas para Encuestas Sociodemográficas).
    *   *Nota metodológica:* Es una variable de hasta 4 dígitos con cientos de categorías específicas (ej. "Fabricación de calzado", "Venta al por menor"). Para tu modelo, te recomiendo fuertemente extraer **los primeros 2 dígitos** para agrupar en macro-sectores (ej. Industria, Comercio, Construcción, Enseñanza). Esto evitará la hiper-fragmentación al crear los Entity Embeddings.
*   **`P47T`**: Monto del ingreso total individual. (Numérica continua).
    *   *Nota:* El valor **-9** significa "No sabe / No responde" y debe coercionarse a nulo (NaN) en el preprocesamiento.
